#### Malous manually annotated images

```python
filelist = sorted(os.listdir('/home/fvitez/nuclei_detection/manually_annotated_images/malou/SegmentationClass/'))
#try:
#    filelist.remove('.ipynb_checkpoints')

image_names = [os.path.join('/home/fvitez/nuclei_detection/manually_annotated_images/malou/SegmentationClass/', f) for f in filelist]
```

```python
imagebuffer = skimage.io.imread_collection(image_names)
images = imagebuffer.concatenate()
```

```python
sorted(os.listdir('/home/fvitez/nuclei_detection/aitslab_boundary_labels/'))
```

```python
#folder_list = sorted(os.listdir('/home/fvitez/nuclei_detection/aitslab_boundary_labels/'))
#folder_list.remove('ipynb_checkpoints')



labels_dir = '/home/fvitez/nuclei_detection/aitslab_boundary_labels/'


for n,i in enumerate(images):

    # strip the first channel
    if len(i.shape) == 3:
        i = i[:,:,0]
    
    annot = skimage.morphology.label(i) # convert to annotation
    boundaries = skimage.segmentation.find_boundaries(annot) #extract boundaries
    
    for k in range(2, 2, 2):
        boundaries = skimage.morphology.binary_dilation(boundaries)
    
    # prepare buffer for binary label
    label_binary = np.zeros((annot.shape + (3,)))

    # write binary label
    label_binary[(annot == 0) & (boundaries == 0), 0] = 1
    label_binary[(annot != 0) & (boundaries == 0), 1] = 1
    label_binary[boundaries == 1, 2] = 1
    
    # save it - converts image to range from 0 to 255
    skimage.io.imsave(labels_dir + filelist[n][:-13] + '/' + filelist[n], label_binary)
```
