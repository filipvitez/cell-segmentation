#### Convert to png & normalize & and save in dir

```python


folder_list = sorted(os.listdir('/home/fvitez/nuclei_detection/tiff_images'))
folder_list.remove('.DS_Store')

for f in folder_list:
    
    flist = sorted(os.listdir('/home/fvitez/nuclei_detection/tiff_images/' + f))
    fdir = '/home/fvitez/nuclei_detection/tiff_images/' + f + "/"

    # run over all raw images
    for filename in tqdm(flist):
        if filename != ".ipynb_checkpoints" and filename != "norm_images" and filename != "boundary_labels":


            # load image and its annotation
            orig_img = skimage.io.imread(fdir + filename, plugin='matplotlib')       

            # IMAGE

            # normalize to [0,1]
            percentile = 99.9
            high = np.percentile(orig_img, percentile)
            low = np.percentile(orig_img, 100-percentile)

            img = np.minimum(high, orig_img)
            img = np.maximum(low, img)

            img = (img - low) / (high - low) # gives float64, thus cast to 8 bit later
            img = skimage.img_as_ubyte(img) 

            skimage.io.imsave('/home/fvitez/nuclei_detection/aitslab_png_images/' + f + '/' + filename[:-4] + 'png', img)
```

#### Reshape and save

```python
folder_list = sorted(os.listdir('/home/fvitez/nuclei_detection/aitslab_png_images/'))
folder_list.remove('.ipynb_checkpoints')

for f in folder_list:
    flist = sorted(os.listdir('/home/fvitez/nuclei_detection/aitslab_png_images/' + f))
    fdir = '/home/fvitez/nuclei_detection/aitslab_png_images/' + f + "/"

    # run over all raw images
    for filename in tqdm(flist):
        if filename != ".ipynb_checkpoints" and filename != "norm_images" and filename != "boundary_labels":


            # load image and its annotation
            orig_img = skimage.io.imread(fdir + filename)       
            i = resize(orig_img, (520, 696), mode='constant', preserve_range=True)
            # IMAGE


            skimage.io.imsave('/home/fvitez/nuclei_detection/resized_aitslab_png_images/' + filename, i)
```
